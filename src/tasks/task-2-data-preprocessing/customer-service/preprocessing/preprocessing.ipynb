```python
for _ , col in enumerate(cols):
    
    #word and char counts
    
    df[f"{col}_word_count"] = df[f"{col}"].apply(lambda x : len(str(x).split()))
    df[f"{col}_char_count"] = df[f"{col}"].apply(lambda x : len(str(x)))
```

```python
def avg_word_length(x):
    
    words = x.split()
    
    len_words = 0
    
    for word in words:
        
        len_words += len(word)
    
    return len_words/len(words)
    
    
```

```python
for col in cols:
    
    #avg word length count
    
    df[f"avg_{col}_word_length"] = df[f"{col}"].apply(lambda x : avg_word_length(x))
```

```python
#contractions to expansions

def con_to_exp(x):
    
    if type(x) is str:
        
        for key in contractions:
            
            val = contractions[key]
            
            x = x.replace(key, val)
     
        return x
    else:
        return x
```

```python
for col in cols:
    
    df[f"{col}_pre"] = df[f"{col}_pre"].apply(lambda x : con_to_exp(str(x)))
```
