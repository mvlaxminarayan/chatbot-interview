```python

section = driver.find_elements(By.CSS_SELECTOR, 'div[class="col"]')
questions_answers = []

for index, sec in enumerate(section):
        if index > 0:
                
            more_responses = sec.find_element(By.CSS_SELECTOR,'a[class="mt-lg d-flex align-items-center"]' )

            driver.execute_script("arguments[0].click();", more_responses)
            
          
        try:
            show_more_responses = sec.find_element(By.CSS_SELECTOR, 'a[class="col d-flex justify-content-center py mb"]')
            
            driver.execute_script("arguments[0].click()",show_more_responses)
        except:
            print(f"there is no more responses to show in question {index+1} ")
            
        
            
        position = sec.find_element(By.CSS_SELECTOR,'a[class="css-1entq9v edupdmz4"]')
        position = position.get_attribute("innerHTML")
        question = sec.find_element(By.CSS_SELECTOR,"h3[class='css-1jvs3tk edupdmz3']")
        question = question.get_attribute("innerHTML")
        responses = sec.find_elements(By.CSS_SELECTOR, "p[class='css-1jbawm2 edupdmz9']")
        questions_answers.append([])
        questions_answers[index].append(position)
        questions_answers[index].append(question)
        for num, response in enumerate(responses):
            
            try:
                expand_response = response.find_element(By.LINK_TEXT, "More")
                driver.execute_script("arguments[0].click()", expand_response)
            
            except:
                print(f"there is no expantion in question {index+1},response {num+1}")
                
            
            response = response.text
            
            questions_answers[index].append(response)

        
print(questions_answers)
```
